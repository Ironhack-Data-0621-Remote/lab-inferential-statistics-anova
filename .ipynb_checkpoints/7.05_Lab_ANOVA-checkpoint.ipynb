### Data import

```python
import pandas as pd
import numpy as np
import math
import scipy.stats as stats
import statsmodels.stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_excel('files_for_lab/anova_lab_data.xlsx', sheet_name='data_collected')
data
```

### Data cleaing

```python
def clean_headers(df):
    df.columns = df.columns.str.lower().str.strip().str.replace(' ', '_')
    return df
```

```python
data = clean_headers(data)
data
```

**H0:**

The mean etching rate is constant irrespective of changes in power of the plasma beam

**HA:** 

There are significant differences in the mean etching rate with changes in the power of the plasma beam

**Significance level:** 0.05

**Degrees of freedom of model:** 2

**Error terms:** 12

**Total DoF:** 14

## Part 2

- In this section, use Python to conduct ANOVA.
- What conclusions can you draw from the experiment and why?

```python
data.groupby('power').agg(np.mean)
```

```python
model = ols('etching_rate ~ C(power)',data=data).fit()
sm.stats.anova_lm(model)
```

**Conclusion:**

The p value of 0.000008 is smaller than the chosen significance level (0.05) so H0 can be rejected. The mean difference of least one of the measured levels of power is significant.
