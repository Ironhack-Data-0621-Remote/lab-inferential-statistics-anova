```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
# read data
data = pd.read_excel('files_for_lab/anova_lab_data.xlsx')# , sheet_name='data_collected')
data
```

```python
data.describe()
```

```python
#rename columns to avoid error when trying to group by "Power"
data.rename(columns = {'Power ':'power','Etching Rate':'etching_rate'}, inplace=True)
```

```python
data.groupby('power').agg(np.mean)
```

Our first intuition is that running 160W, the machine has the lowest mean etching rate.

Testing with ANOVA: <br>
**H0: There is no difference in the mean etching rate for different levels of power <br>
H1: There is a difference in the mean etching rate for different levels of power**

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

model = ols('etching_rate ~ C(power)',data=data).fit()
sm.stats.anova_lm(model)
```

Result: 0.000008 < 0.05 means that H0 can be rejected. There is a significant difference between applying different levels of power.
